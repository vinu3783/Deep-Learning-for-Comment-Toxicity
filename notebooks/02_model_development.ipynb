```python
# Step 2: Memory-Optimized Model Development for Toxicity Detection
# ==================================================================
# Project Path: D:\Deep Learning for Comment Toxicity

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import pickle
import gc
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.feature_extraction.text import TfidfVectorizer

# Deep Learning imports
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LSTM, GRU, Embedding, Dropout, GlobalMaxPooling1D, Conv1D, MaxPooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras import backend as K

warnings.filterwarnings('ignore')
plt.style.use('default')

# Memory optimization settings
tf.config.experimental.enable_memory_growth = True
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

def clear_memory():
    """Clear memory and garbage collect"""
    K.clear_session()
    gc.collect()

print("Libraries imported successfully!")
print(f"TensorFlow version: {tf.__version__}")

# ================================================================
# 1. LOAD PREPROCESSED DATA WITH MEMORY OPTIMIZATION
# ================================================================

print("\n=== LOADING PREPROCESSED DATA ===")

# Load preprocessed datasets
try:
    train_processed = pd.read_csv('../data/processed/train_processed.csv')
    test_processed = pd.read_csv('../data/processed/test_processed.csv')
    print(f"Train processed shape: {train_processed.shape}")
    print(f"Test processed shape: {test_processed.shape}")
except FileNotFoundError:
    print("Error: Preprocessed data not found. Please run Step 1 first.")
    raise

# MEMORY OPTIMIZATION: Sample data if too large
if len(train_processed) > 50000:
    print(f"Large dataset detected ({len(train_processed)} samples). Sampling for memory efficiency...")
    train_processed = train_processed.sample(n=50000, random_state=42).reset_index(drop=True)
    print(f"Sampled dataset size: {len(train_processed)}")

# Define toxicity columns
toxicity_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

# Check data
print("\nDataset info:")
print(f"Training samples: {len(train_processed)}")
print(f"Toxicity distribution:")
for col in toxicity_columns:
    print(f"  {col}: {train_processed[col].sum()} ({(train_processed[col].sum()/len(train_processed)*100):.1f}%)")

# ================================================================
# 2. MEMORY-EFFICIENT TEXT TOKENIZATION
# ================================================================

print("\n=== PREPARING TEXT SEQUENCES (MEMORY OPTIMIZED) ===")

# Prepare texts and labels
texts = train_processed['comment_text_processed'].fillna('').astype(str)
labels = train_processed[toxicity_columns].values

# REDUCED parameters for memory efficiency
max_features = 10000  # Reduced from 20000
max_length = 50       # Reduced from 100

# Create and fit tokenizer
tokenizer = Tokenizer(num_words=max_features, oov_token='<OOV>')
tokenizer.fit_on_texts(texts)

# Convert texts to sequences
sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')

print(f"Vocabulary size: {len(tokenizer.word_index)}")
print(f"Sequences shape: {X.shape}")
print(f"Labels shape: {labels.shape}")

# Prepare test data
test_texts = test_processed['comment_text_processed'].fillna('').astype(str)
test_sequences = tokenizer.texts_to_sequences(test_texts)
X_test = pad_sequences(test_sequences, maxlen=max_length, padding='post', truncating='post')

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(
    X, labels, test_size=0.2, random_state=42, stratify=train_processed['is_toxic']
)

print(f"Training set shape: {X_train.shape}")
print(f"Validation set shape: {X_val.shape}")
print(f"Test set shape: {X_test.shape}")

# Clean up memory
del texts, sequences, labels
gc.collect()

# ================================================================
# 3. BASELINE MODEL - LOGISTIC REGRESSION
# ================================================================

print("\n=== BUILDING BASELINE MODEL (TF-IDF + LOGISTIC REGRESSION) ===")

# Create TF-IDF features with reduced parameters
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1, 2))
X_tfidf_train = tfidf.fit_transform(train_processed['comment_text_processed'].fillna(''))

# Split TF-IDF data
X_tfidf_train_split, X_tfidf_val_split, y_tfidf_train, y_tfidf_val = train_test_split(
    X_tfidf_train, train_processed[toxicity_columns].values, test_size=0.2, random_state=42, 
    stratify=train_processed['is_toxic']
)

# Train baseline model
baseline_model = MultiOutputClassifier(LogisticRegression(random_state=42, max_iter=500))
baseline_model.fit(X_tfidf_train_split, y_tfidf_train)

# Evaluate baseline
baseline_pred = baseline_model.predict(X_tfidf_val_split)
baseline_pred_proba = baseline_model.predict_proba(X_tfidf_val_split)

print("Baseline Model Performance:")
baseline_auc_scores = []
for i, col in enumerate(toxicity_columns):
    auc = roc_auc_score(y_tfidf_val[:, i], baseline_pred_proba[i][:, 1])
    acc = accuracy_score(y_tfidf_val[:, i], baseline_pred[:, i])
    baseline_auc_scores.append(auc)
    print(f"  {col}: AUC={auc:.3f}, Accuracy={acc:.3f}")

baseline_mean_auc = np.mean(baseline_auc_scores)
print(f"  Mean AUC: {baseline_mean_auc:.3f}")

# Clean up memory
del X_tfidf_train, X_tfidf_train_split, X_tfidf_val_split
gc.collect()

# ================================================================
# 4. MEMORY-EFFICIENT DEEP LEARNING MODELS
# ================================================================

# Model parameters (reduced for memory efficiency)
embedding_dim = 50    # Reduced from 100
vocab_size = min(max_features, len(tokenizer.word_index)) + 1
batch_size = 16       # Reduced from 32
epochs = 5            # Reduced from 10

# Create models directory
os.makedirs('../models', exist_ok=True)

# Store model results
models_results = {}

# ================================================================
# MODEL 1: SIMPLE NEURAL NETWORK (MEMORY OPTIMIZED)
# ================================================================

print("\n=== BUILDING MODEL 1: SIMPLE NEURAL NETWORK ===")

def create_simple_nn(vocab_size, embedding_dim, max_length, num_classes):
    model = Sequential([
        Embedding(vocab_size, embedding_dim, input_length=max_length),
        GlobalMaxPooling1D(),
        Dense(64, activation='relu'),  # Reduced from 128
        Dropout(0.5),
        Dense(32, activation='relu'),  # Reduced from 64
        Dropout(0.3),
        Dense(num_classes, activation='sigmoid')
    ])
    return model

# Clear memory before creating model
clear_memory()

# Create and compile model
model_simple = create_simple_nn(vocab_size, embedding_dim, max_length, len(toxicity_columns))
model_simple.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("Simple Neural Network Architecture:")
model_simple.summary()

# Callbacks
callbacks_simple = [
    EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, min_lr=1e-6)
]

# Train model
print("Training Simple Neural Network...")
history_simple = model_simple.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X_val, y_val),
    callbacks=callbacks_simple,
    verbose=1
)

# Evaluate model
simple_pred = model_simple.predict(X_val, batch_size=batch_size)
simple_auc_scores = []
print("Simple NN Performance:")
for i, col in enumerate(toxicity_columns):
    auc = roc_auc_score(y_val[:, i], simple_pred[:, i])
    pred_binary = (simple_pred[:, i] > 0.5).astype(int)
    acc = accuracy_score(y_val[:, i], pred_binary)
    simple_auc_scores.append(auc)
    print(f"  {col}: AUC={auc:.3f}, Accuracy={acc:.3f}")

simple_mean_auc = np.mean(simple_auc_scores)
models_results['Simple NN'] = {'auc': simple_mean_auc, 'history': history_simple}
print(f"  Mean AUC: {simple_mean_auc:.3f}")

# Save model
model_simple.save('../models/simple_nn_model.h5')

# Clear memory
del model_simple, simple_pred
clear_memory()

# ================================================================
# MODEL 2: LSTM (MEMORY OPTIMIZED)
# ================================================================

print("\n=== BUILDING MODEL 2: MEMORY-OPTIMIZED LSTM ===")

def create_lstm_model(vocab_size, embedding_dim, max_length, num_classes):
    model = Sequential([
        Embedding(vocab_size, embedding_dim, input_length=max_length),
        LSTM(32, dropout=0.3, recurrent_dropout=0.3),  # Much smaller LSTM
        Dense(32, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='sigmoid')
    ])
    return model

# Clear memory before creating model
clear_memory()

# Create and compile LSTM model
model_lstm = create_lstm_model(vocab_size, embedding_dim, max_length, len(toxicity_columns))
model_lstm.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("LSTM Model Architecture:")
model_lstm.summary()

# Callbacks
callbacks_lstm = [
    EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, min_lr=1e-6)
]

# Train LSTM model
print("Training LSTM Model...")
history_lstm = model_lstm.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X_val, y_val),
    callbacks=callbacks_lstm,
    verbose=1
)

# Evaluate model
lstm_pred = model_lstm.predict(X_val, batch_size=batch_size)
lstm_auc_scores = []
print("LSTM Performance:")
for i, col in enumerate(toxicity_columns):
    auc = roc_auc_score(y_val[:, i], lstm_pred[:, i])
    pred_binary = (lstm_pred[:, i] > 0.5).astype(int)
    acc = accuracy_score(y_val[:, i], pred_binary)
    lstm_auc_scores.append(auc)
    print(f"  {col}: AUC={auc:.3f}, Accuracy={acc:.3f}")

lstm_mean_auc = np.mean(lstm_auc_scores)
models_results['LSTM'] = {'auc': lstm_mean_auc, 'history': history_lstm}
print(f"  Mean AUC: {lstm_mean_auc:.3f}")

# Save model
model_lstm.save('../models/lstm_model.h5')

# Clear memory
del model_lstm, lstm_pred
clear_memory()

# ================================================================
# MODEL 3: CNN (MEMORY OPTIMIZED)
# ================================================================

print("\n=== BUILDING MODEL 3: MEMORY-OPTIMIZED CNN ===")

def create_cnn_model(vocab_size, embedding_dim, max_length, num_classes):
    model = Sequential([
        Embedding(vocab_size, embedding_dim, input_length=max_length),
        Conv1D(64, 3, activation='relu'),  # Reduced filters
        MaxPooling1D(2),
        Conv1D(32, 3, activation='relu'),  # Reduced filters
        GlobalMaxPooling1D(),
        Dense(32, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='sigmoid')
    ])
    return model

# Clear memory before creating model
clear_memory()

# Create and compile CNN model
model_cnn = create_cnn_model(vocab_size, embedding_dim, max_length, len(toxicity_columns))
model_cnn.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("CNN Model Architecture:")
model_cnn.summary()

# Callbacks
callbacks_cnn = [
    EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, min_lr=1e-6)
]

# Train CNN model
print("Training CNN Model...")
history_cnn = model_cnn.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X_val, y_val),
    callbacks=callbacks_cnn,
    verbose=1
)

# Evaluate model
cnn_pred = model_cnn.predict(X_val, batch_size=batch_size)
cnn_auc_scores = []
print("CNN Performance:")
for i, col in enumerate(toxicity_columns):
    auc = roc_auc_score(y_val[:, i], cnn_pred[:, i])
    pred_binary = (cnn_pred[:, i] > 0.5).astype(int)
    acc = accuracy_score(y_val[:, i], pred_binary)
    cnn_auc_scores.append(auc)
    print(f"  {col}: AUC={auc:.3f}, Accuracy={acc:.3f}")

cnn_mean_auc = np.mean(cnn_auc_scores)
models_results['CNN'] = {'auc': cnn_mean_auc, 'history': history_cnn}
print(f"  Mean AUC: {cnn_mean_auc:.3f}")

# Save model
model_cnn.save('../models/cnn_model.h5')

# Clear memory
del model_cnn, cnn_pred
clear_memory()

# ================================================================
# MODEL COMPARISON AND VISUALIZATION
# ================================================================

print("\n=== MODEL COMPARISON ===")

# Add baseline to results
models_results['Baseline'] = {'auc': baseline_mean_auc, 'history': None}

# Find best model
best_model_name = max(models_results.keys(), key=lambda x: models_results[x]['auc'])
best_auc = models_results[best_model_name]['auc']

print("Model Performance Summary:")
for name, results in models_results.items():
    print(f"  {name}: AUC={results['auc']:.3f}")

print(f"\nBest model: {best_model_name} with AUC: {best_auc:.3f}")

# Plot training histories (exclude baseline)
histories = {name: results['history'] for name, results in models_results.items() 
            if results['history'] is not None}

if histories:
    plt.figure(figsize=(15, 5))
    
    for i, (name, history) in enumerate(histories.items()):
        plt.subplot(1, len(histories), i+1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{name} - Training History')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Compare model performances
plt.figure(figsize=(10, 6))
model_names = list(models_results.keys())
aucs = [models_results[name]['auc'] for name in model_names]
bars = plt.bar(model_names, aucs, color='skyblue', alpha=0.7)
plt.title('Model Comparison - Mean AUC')
plt.ylabel('AUC Score')
plt.xticks(rotation=45)
plt.ylim(0, 1)

# Add value labels on bars
for bar, auc in zip(bars, aucs):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{auc:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# ================================================================
# SAVE BEST MODEL AND GENERATE PREDICTIONS
# ================================================================

print("\n=== SAVING BEST MODEL AND GENERATING PREDICTIONS ===")

# Load the best model if it's not baseline
if best_model_name != 'Baseline':
    best_model = tf.keras.models.load_model(f'../models/{best_model_name.lower().replace(" ", "_")}_model.h5')
    
    # Generate predictions for test set
    test_predictions = best_model.predict(X_test, batch_size=batch_size)
    
    # Save best model with standard name
    best_model.save('../models/best_toxicity_model.h5')
    print(f"Best model ({best_model_name}) saved as: best_toxicity_model.h5")
else:
    print("Baseline model is best - using TF-IDF approach for predictions")
    # For baseline, use TF-IDF on test data
    X_test_tfidf = tfidf.transform(test_processed['comment_text_processed'].fillna(''))
    test_predictions = baseline_model.predict_proba(X_test_tfidf)
    test_predictions = np.column_stack([pred[:, 1] for pred in test_predictions])

# Save tokenizer and parameters
with open('../models/tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

model_params = {
    'max_features': max_features,
    'max_length': max_length,
    'embedding_dim': embedding_dim,
    'vocab_size': vocab_size,
    'toxicity_columns': toxicity_columns,
    'best_model_name': best_model_name,
    'best_model_auc': best_auc
}

with open('../models/model_params.pickle', 'wb') as handle:
    pickle.dump(model_params, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Create submission dataframe
submission_df = test_processed[['id']].copy()
for i, col in enumerate(toxicity_columns):
    submission_df[col] = test_predictions[:, i]

submission_df.to_csv('../data/processed/test_predictions.csv', index=False)
print("Test predictions saved as: test_predictions.csv")

# ================================================================
# SUMMARY REPORT
# ================================================================

print("\n" + "="*60)
print("MEMORY-OPTIMIZED MODEL DEVELOPMENT SUMMARY")
print("="*60)
print(f"Dataset size used: {len(train_processed)} samples")
print(f"Models trained: {len([k for k in models_results.keys() if k != 'Baseline'])}")
print(f"Best performing model: {best_model_name}")
print(f"Best model AUC: {best_auc:.3f}")
print(f"Memory optimizations applied:")
print(f"  - Reduced vocabulary: {max_features}")
print(f"  - Reduced sequence length: {max_length}")
print(f"  - Reduced embedding dimension: {embedding_dim}")
print(f"  - Reduced batch size: {batch_size}")
print(f"  - Reduced epochs: {epochs}")
print(f"  - Memory cleanup between models")

print("\nModel Performance Summary:")
for name, results in models_results.items():
    print(f"  {name}: AUC={results['auc']:.3f}")

print("\nFiles saved:")
print(f"  - Best model: ../models/best_toxicity_model.h5")
print(f"  - Tokenizer: ../models/tokenizer.pickle")
print(f"  - Model parameters: ../models/model_params.pickle")
print(f"  - Test predictions: ../data/processed/test_predictions.csv")
print("="*60)

print("\nStep 2 (Memory-Optimized Model Development) completed successfully!")
print("Ready for Step 3: Streamlit Application Development")
```
